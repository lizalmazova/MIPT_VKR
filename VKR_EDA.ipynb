# Визуализация

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df, x='Crisis')
plt.title('Cоотношение исходов')
plt.xlabel('Статус кризиса')
plt.ylabel('Количество исходов')

# Добавление аннотаций (меток) на столбцах графика
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Создание круговой диаграммы с помощью Matplotlib и Seaborn
labels = ['Stable', 'Crisis period']
sizes = df['Crisis'].value_counts()
#colors = ['skyblue', 'black']

plt.figure(figsize=(5, 4))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Количество исходов')
plt.axis('equal')  # Установка равного масштаба для круговой диаграммы

# Отображение диаграммы
plt.show()
```

```python
labels=['Stable', 'Crisis period']
import plotly.express as px
fig = px.pie(values=df['Crisis'].value_counts(), names=labels , width=700, height=400, color_discrete_sequence=["skyblue","black"]
             ,title="Количество исходов")
fig.show()
```

```python
import plotly.express as px


# Создание столбчатой диаграммы с помощью Plotly Express
fig = px.histogram(df, x='Crisis', title='Cоотношение исходов',
                   labels={'Crisis': 'Статус кризиса'},
                   category_orders={'Crisis': df['Crisis'].value_counts().index},
                   barmode='relative')

# Настройка осей и сетки (опционально)
fig.update_xaxes(title='Статус кризиса')
fig.update_yaxes(title='Количество исходов')
fig.update_layout(showlegend=False)  # Отключение легенды (если не требуется)
fig.update_layout(xaxis=dict(showgrid=True, gridwidth=1, gridcolor='LightGray'),
                  yaxis=dict(showgrid=True, gridwidth=1, gridcolor='LightGray'))

# Отображение графика
fig.show()
```

```python
from matplotlib import pyplot as plt
def boxplots_custom(dataset, columns_list, rows, cols, suptitle):
    fig, axs = plt.subplots(rows, cols, sharey=True, figsize=(10,7))
    fig.suptitle(suptitle,y=1, size=25)
    axs = axs.flatten()
    for i, data in enumerate(columns_list):
        sns.boxplot(data=dataset[data], orient='h', ax=axs[i])
        axs[i].set_title(data + ', skewness is: '+str(round(dataset[data].skew(axis = 0, skipna = True),2)))

boxplots_custom(dataset=df, columns_list=(list(df.loc[:, 'Var1':'Var9'])), rows=3, cols=3, suptitle='Boxplots for each variable')
plt.tight_layout()
```
